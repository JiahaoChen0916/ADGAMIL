```python
import torch
from AttMIL.network import DAttention

x = torch.randn(1, 500, 1024).cuda()
mask = torch.ones(1, 500).bool().cuda()

model = DAttention(n_classes=2, dropout=0.25, act='relu', n_features=1024).cuda()
logits, Y_prob, Y_hat, A_raw, results_dict = model(x, attn_mask=mask)

print(logits.shape)   # [1, 2]
print(Y_prob.shape)   # [1, 2]
print(Y_hat.shape)    # [1, 1]
print(A_raw.shape)    # [1, 1, 500]
print(results_dict)   # {}
```
